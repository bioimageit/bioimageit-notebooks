# bioimageit_core: install

In this tutorial, we will show several methods to install `bioimageit_core`. `bioimageit_core` is a python library written in *python3*. It is a python API that allows to do all the BioImageIT data manipulation: 1- generate and manipulate metadata files for raw and processed data and 2- run processes on taged data.

As a *python3* library, `bioimageit_core` can be installed by 3 methods: 
1. Using *Conda*
3. Using *pip*
2. Add the library to the path

## Using Conda

Using *Conda* is the recomended method to instal `bioimageit_core` since most of the BioImageIT tools dependencies are managed with *Conda*.

To install `bioimageit_core`, using *Conda*, type the following commands.
First we recommand using a environement:
```
conda create --name bioimageit
conda activate bioimageit
```
and then install `bioimageit_core`
```
conda install -c bioimageit bioimageit_core
```

## Install using *pip*

*pip* is the python package manager. To install a package with *pip* you just need to run the command `python3 -m pip install SomePackage` 
You can then install `bioimageit_core` using the command:
```
python -m pip install bioimageit_core
```
